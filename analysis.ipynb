## -- What are the top 3 hotspots in each burough by count?

NYC has 5 different buroughs so we will generate clusters of Airbnbs based on count in every neighbourhood of each burough. 

The clusters will be looked at for the year of 2019 which is the last year of the dataset with the assumption that by this year the clusters will be well defined.

```python
# Taking the latest year of the Airbnb data
val = 2019
```

Each Airbnb listing has its own latitude and longitude and therefore,  in order to create a cluster for a particular neighbourhood we need a standard latitude and longitude for that neighbourhood. 

In order to do this, an average of the latitude and longitutde will be computed and imputed into the listings data on the basis of the  neighbourhood the Airbnb belongs to. This will help in showing the count (cluster) of Airbnb listings in each neighbourhood.

```python
# Finding the mean lat_long of Airbnbs in a particular neighbourhood to standardize the map plot
airbnb_neigh_lat_long = data_airbnb.groupby('neighbourhood', as_index=False)['latitude','longitude'].mean()
airbnb_neigh_lat_long
```

```python
# Imputing the average lat_long for each neighbourhood
data_airbnb['latitude_impute'] = data_airbnb['neighbourhood'].map(airbnb_neigh_lat_long.set_index('neighbourhood')['latitude'])
data_airbnb['longitude_impute'] = data_airbnb['neighbourhood'].map(airbnb_neigh_lat_long.set_index('neighbourhood')['longitude'])
```

```python
# Filtering all the data till input year (inclusive)
data_airbnb_2019 = data_airbnb.loc[data_airbnb['host_since_year'] < (val+ 1)].reset_index(drop = True)
data_airbnb_2019.head()
```

```python
#Getting the Airbnb count by neighbourhood
counts_airbnb = data_airbnb_2019.neighbourhood.value_counts()
counts_airbnb = pd.DataFrame({'neighbourhood':counts_airbnb.index, 'count':counts_airbnb.values})
counts_airbnb
```

```python
# Grouping Airbnb counts by neighbourhood to show the Airbnb clusters on a map (count = size)
df_temp = data_airbnb_2019.groupby('neighbourhood',as_index=False)['neighbourhood'].count()
df_map = data_airbnb_2019.groupby('neighbourhood',as_index=False)['neighbourhood', 'borough','latitude_impute', 'longitude_impute'].max()

df_map['count'] = df_temp['neighbourhood']
```

```python
# Plotting the airbnb clusters on a map (by size)
fig1 = px.scatter_mapbox(df_map, lat="latitude_impute", lon="longitude_impute", size = 'count', color = 'borough', 
                        hover_name = 'neighbourhood', title = 'Airbnb clusters in NYC neighbourhoods',
                  color_continuous_scale=px.colors.cyclical.IceFire,width=700, height=800, size_max=50,zoom=10)

fig1.show()
```

## -- What is the growth of rate of these hotspots?

```python
#Converting the top neighbourhoods to a list in order to do a year-on-year growth analysis
top_neighbourhoods = df_map2['neighbourhood'].tolist()
top_neighbourhoods
```

```python
#Filtering the airbnb data for the top neighbourhoods
data_topneigh_airbnb = data_airbnb[data_airbnb['neighbourhood'].isin(top_neighbourhoods)]
```

```python
# Defining a function to get a count of listings each year from the dataset for these top neighbourhoods
def getcountbyyear(df):
  '''
  The function takes the dataframe and returns the dataframe with the count per year and total count for each neighborhood
  '''
  for year in range(df['host_since_year'].min(),df['host_since_year'].max()):
      data_airbnb_byyear = data_topneigh_airbnb.loc[data_topneigh_airbnb['host_since_year'] < (year+ 1)].reset_index(drop = True)
      data_airbnb_byyear.head()
        
      counts_airbnb = data_airbnb_byyear.neighbourhood.value_counts()
      counts_airbnb = pd.DataFrame({'neighbourhood':counts_airbnb.index, 'count':counts_airbnb.values})
      counts_airbnb

      df_temp = data_airbnb_byyear.groupby(['neighbourhood','host_since_year'], as_index=False)['neighbourhood','host_since_year'].size()
      df_final = data_airbnb_byyear.groupby(['neighbourhood','host_since_year'], as_index=False)['neighbourhood', 'borough',
                                             'latitude_impute', 'longitude_impute','host_since_year'].max()
      df_final['count_per_year'] = df_temp['size']
      df_final['total_count'] = df_final.groupby(['neighbourhood'])['count_per_year'].cumsum()

  return df_final
```

```python
hotspot_lat = df_topneighborhoods_final.latitude_impute.unique().tolist()
hotspot_lon = df_topneighborhoods_final.longitude_impute.unique().tolist()
```

```python
def filter_latlon (df,latlist,lonlist):
    '''
    The function takes the crime dataframe and list of latitudes and longituted for airbnb hotspots and the filters the crime dataframe
    for crimes within one km radius of the airbnb hotspots 
    '''
    latlon = list(zip(latlist,lonlist))
    df_result = df[0:0]
    for i,j in latlon:
        df_temp = df[df['latitude'].between(i-0.01,i+0.01)]
        df_temp2 = df_temp[df_temp['longitude'].between(j-0.01,j+0.01)]
        df_result = pd.concat([df_result, df_temp2], axis=0)

    return df_result
    
```

# 3) If the Airbnb listing description includes the word safe, what is the average rating in comparison to when it is not mentioned?

## Is the difference statistically significant?

```python
# Creating a dataframe for listings that mention safe in the description
df_safe = data_airbnb[data_airbnb['description'].str.contains(' safe ')]
```

```python
df_safe['rating'].hist()
```

The ratings data is left skewed indicating that most tourists rate the Airbnb listings highly. Now the interesting thing to look at here would be the ratings that are two standard deviations below the mean rating and try to see if safety was a concern here even after safe was mentioned in the description.

```python
df_safe['rating'].describe()
```

```python
ratings_lower_bound = 93.387806 - 2*9.192678
ratings_lower_bound
```

Filtering the Airbnb listings with low ratings

```python
df_safe_lower_ratings = df_safe.loc[df_safe['rating'] < ratings_lower_bound]   
df_safe_lower_ratings['rating'].describe()
```

The average rating was 58.28 and the standard deviation was 16.09 with 109 listings.

```python
# Creating a dataframe for listings where safe is not mentioned
df_safe_not_mentioned = data_airbnb[~data_airbnb['description'].str.contains(' safe ')]
```

```python
df_safe_not_mentioned['rating'].describe()
```

```python
ratings_lower_bound = 93.876861 - 2*9.221801
ratings_lower_bound
```

```python
df_lower_ratings = df_safe_not_mentioned.loc[df_safe_not_mentioned['rating'] < ratings_lower_bound]   
df_lower_ratings['rating'].describe()
```

The average rating was 56.93 and the standard deviation was 17.17 with 1113 listings.

Now let's see if this difference is statistically significant. 

The null hypothesis is that mentioning the word 'safe' in the description does not affect the rating.

The alternate hypothesis is that mentioning the word 'safe' in the description leads to a higher rating.


**The code below was used from an online article published by Chao De-Yu**

**Article Name: Two-Sample Hypothesis Tests, with Python**

**Article Link: https://levelup.gitconnected.com/two-sample-hypothesis-tests-with-python-43e1b8c52306**

```python
alpha = 0.1
safe_ratings = np.array(df_safe_lower_ratings['rating'])
safe_not_mentioned_ratings = np.array(df_lower_ratings['rating'])

# possible types "right-tailed, left-tailed, two-tailed"
tail_hypothesis_type = "two-tailed"

if tail_hypothesis_type == "two-tailed":
    alpha /= 2

# calculate number of sample for both female and male
n1 = len(safe_ratings)
n2 = len(safe_not_mentioned_ratings)

# calculate sample mean
x1_bar = np.mean(safe_ratings)
x2_bar = np.mean(safe_not_mentioned_ratings)

# calculate sample standard deviation => ddof = 1, population std => ddof = 0
s1 = np.std(safe_ratings, ddof=1)
s2 = np.std(safe_not_mentioned_ratings, ddof=1)

print("Basic Statistics from sample data")
print("Number of safe_ratings sample:", n1, "\nNumber of safe_not_mentioned_ratings sample:", n2)
print("\nsafe_ratings mean:", x1_bar, "\nsafe_not_mentioned_ratings mean:", x2_bar)
print("\nsafe_ratings std:", s1, "\nsafe_not_mentioned_ratings std:", s2)

# Test statistics, on upper critical value
if s1 > s2:
    F_score = s1**2 / s2**2
    critical_value = stats.f.ppf(1-alpha, n1-1, n2-1)
else:
    F_score = s2**2 / s1**2
    critical_value = stats.f.ppf(1-alpha, n2-1, n1-1)

print("\n----------------------------TEST STATISTICS---------------------------------------------")
print("F-score is:", F_score, " and critical value is:", critical_value)
conclusion = "Failed to reject the null hypothesis."
t_test_type = "pooled_variance t test"
if F_score > critical_value:
    conclusion = "Null Hypothesis is rejected."
    t_test_type = "separate-variances t test"
print(conclusion)
```

#### **Analysis for whether the word safe has an impact on listing rating**

From the analysis above, it can be seen that while the ratings are higher for listings that mention the word 'safe' in their description, that value is not statistically significant when compared with listings that do not mention the word 'safe'.

A next step for this analysis would be to develop a Machine Learning model that uses the description information, ratings information and also, the crimes information and labels listings as safe or no. This is a feature (**safety verified by Airbnb**) that Airbnb could implement on their portal to help tourists get a better sense of the neighbourhood in which the Airbnb is located.
